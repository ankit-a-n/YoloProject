```python

from google.colab import files
import cv2
from ultralytics import YOLO, solutions
import matplotlib.pyplot as plt
import os

print("="*60)
print("YOLO-26 CAN DETECT THE FOLLOWING 80 OBJECT CLASSES:")
print("="*60)

model_for_names = YOLO("yolo26n.pt")
class_names = model_for_names.names

for class_id, name in sorted(class_names.items()):
    print(f"{class_id:2}: {name}")
print("="*60)
print("\nWhen setting the 'region_points', remember that objects crossing this line will be counted.\n")

print("Please upload an IMAGE (.jpg, .png) or a VIDEO (.mp4, .avi):")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
file_extension = os.path.splitext(filename)[1].lower()

is_image = file_extension in ('.png', '.jpg', '.jpeg', '.webp')

print(f"\nUploaded file: {filename}")
print(f"File type detected: {'Image' if is_image else 'Video'}")

model = YOLO("yolo26n.pt")

region_points = [(100, 400), (1000, 400)]

counter = solutions.ObjectCounter(
    show=False,
    region=region_points,
    model="yolo26n.pt",

    conf=0.25
)

if is_image:
    print("Processing image...")
    image = cv2.imread(filename)
    results = counter(image)

    plt.figure(figsize=(15, 10))
    plt.imshow(cv2.cvtColor(results.plot_im, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    output_file = "output_counted_image.jpg"
    cv2.imwrite(output_file, results.plot_im)
    files.download(output_file)
    print("Image processing complete. File downloaded.")

else:
    print("Processing video... This may take a few minutes.")
    cap = cv2.VideoCapture(filename)
    assert cap.isOpened(), "Error: Could not read video file."

    w, h, fps = (int(cap.get(x)) for x in (
        cv2.CAP_PROP_FRAME_WIDTH,
        cv2.CAP_PROP_FRAME_HEIGHT,
        cv2.CAP_PROP_FPS
    ))

    output_file = "output_counted_video.mp4"
    video_writer = cv2.VideoWriter(
        output_file,
        cv2.VideoWriter_fourcc(*"mp4v"),
        fps,
        (w, h)
    )

    frame_count = 0
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        results = counter(frame)
        video_writer.write(results.plot_im)
        frame_count += 1

        if frame_count % 50 == 0:
            print(f"  Processed {frame_count} frames...")

    cap.release()
    video_writer.release()
    print(f"\nVideo processing complete. Total frames: {frame_count}")
    print("Downloading the output video...")
    files.download(output_file)

print("\n===== PROJECT EXECUTION FINISHED =====")
```
